```python
import pandas as pd
print(pd.__version__)
```

```python
file_path = input("Enter the full path to your CSV file: ")
```

```python
# Read the CSV file

# The CSV must have at least two columns: "Category" and "Amount"
df = pd.read_csv(file_path)

# Preview the data (optional)
print(df.head())
```

```python
#Clean data

if 'Amount' in df.columns:
    # Convert all values to strings for uniform processing
    df['Amount'] = (
        df['Amount']
        .astype(str)
        .replace(r'[^0-9.\+-]', '', regex=True)  # remove everything except digits, . + and -
        .replace('', '0')                       # replace blanks with 0
    )

    # Convert to numeric, forcing errors to NaN
    df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

    # Drop rows where Amount couldn't be converted
    df = df.dropna(subset=['Amount'])
else:
    print("No'Amount' column.")
    exit()

# Summarize by category
if 'Category' in df.columns and 'Amount' in df.columns:
    category_summary = (
    df.groupby('Category')['Amount']
    .sum()
    .abs()  # make totals positive for display
    .sort_values(ascending=False)
)

    # Print the summary
    print("\nðŸ’° Total spending (or income) by category:")
    print(category_summary)
```
